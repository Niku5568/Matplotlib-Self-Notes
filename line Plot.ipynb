=============    ===============================
character        description
=============    ===============================

=============    ===============================

Example format strings::

    'b'    # blue markers with default shape
    'or'   # red circles
    '-g'   # green solid line
    '--'   # dashed line with default color
    '^k:'  # black triangle_up markers connected by a dotted line

```python
x=[1,2,3,7]
y=[1,7,9,8]
plt.plot(x,y,color="r",linewidth=3,marker="d",markersize=10,markerfacecolor="b",markeredgecolor="b",linestyle="dotted")
plt.title("Line Plot",fontsize=20,loc="left")
plt.xlabel("X - Axis",fontsize=15,loc="right")
plt.ylabel("Y - Axis",fontsize=15,loc="bottom")
plt.show()
```

# change X and Y limit using axis Method

    ====================
    Call signatures
    ====================
    xmin, xmax, ymin, ymax = axis()               
    xmin, xmax, ymin, ymax = axis([xmin, xmax, ymin, ymax])               
    xmin, xmax, ymin, ymax = axis(option)                 
    xmin, xmax, ymin, ymax = axis(**kwargs)                  

    ================ ===========================================================
    Value            Description
    ================ ===========================================================
    'off' or `False` Hide all axis decorations, i.e. axis labels, spines,
                     tick marks, tick labels, and grid lines.
                     This is the same as `~.Axes.set_axis_off()`.
    'on' or `True`   Do not hide all axis decorations, i.e. axis labels, spines,
                     tick marks, tick labels, and grid lines.
                     This is the same as `~.Axes.set_axis_on()`.
    'equal'          Set equal scaling (i.e., make circles circular) by
                     changing the axis limits. This is the same as
                     ``ax.set_aspect('equal', adjustable='datalim')``.
                     Explicit data limits may not be respected in this case.
    'scaled'         Set equal scaling (i.e., make circles circular) by
                     changing dimensions of the plot box. This is the same as
                     ``ax.set_aspect('equal', adjustable='box', anchor='C')``.
                     Additionally, further autoscaling will be disabled.
    'tight'          Set limits just large enough to show all data, then
                     disable further autoscaling.
    'auto'           Automatic scaling (fill plot box with data).
    'image'          'scaled' with axis limits equal to data limits.
    'square'         Square plot; similar to 'scaled', but initially forcing
                     ``xmax-xmin == ymax-ymin``.
    ================ ===========================================================




```python
print(x)
print(y)
plt.plot(x ,color="r",linewidth=3,marker="d",markersize=10,markerfacecolor="b",markeredgecolor="b",linestyle="dotted",label="Line 1")
plt.plot(y,color="b",linewidth=3,marker="o",markersize=10,markerfacecolor="b",markeredgecolor="b",linestyle="dotted",label="Line 2")
plt.title("Line Plot",fontsize=20,loc="left")
plt.xlabel("X - Axis",fontsize=15,loc="right")
plt.ylabel("Y - Axis",fontsize=15,loc="bottom")
# xmin, xmax, ymin, ymax
plt.axis([0,20,0,15])
plt.legend()
plt.show()
```

# change X and Y limit using Xticks and Yticks

```python
x_ticks=[i for i in range(0,20,2)]
y_ticks=[i for i in range(0,15,3)]
print(x_ticks)
print(y_ticks)
```

```python
print(x)
print(y)
plt.plot(x ,color="r",linewidth=3,marker="d",markersize=10,markerfacecolor="b",markeredgecolor="b",linestyle="dotted",label="Line 1")
plt.plot(y,color="b",linewidth=3,marker="o",markersize=10,markerfacecolor="b",markeredgecolor="b",linestyle="dotted",label="Line 2")
plt.title("Line Plot",fontsize=20,loc="left")
plt.xlabel("X - Axis",fontsize=15,loc="right")
plt.ylabel("Y - Axis",fontsize=15,loc="bottom")
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.show()
```

# Use grid -> change grid color -> change grid line

```python
print(x)
print(y)
plt.plot(x ,color="r",linewidth=3,marker="d",markersize=10,markerfacecolor="b",markeredgecolor="b",linestyle="dotted",label="Line 1")
plt.plot(y,color="b",linewidth=3,marker="o",markersize=10,markerfacecolor="b",markeredgecolor="b",linestyle="dotted",label="Line 2")
plt.title("Line Plot",fontsize=20,loc="left")
plt.xlabel("X - Axis",fontsize=15,loc="right")
plt.ylabel("Y - Axis",fontsize=15,loc="bottom")
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color="orange",linestyle="--",linewidth=1)
plt.show()
```
