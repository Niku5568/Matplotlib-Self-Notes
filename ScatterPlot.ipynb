# scatter Plot -> used to show relationship b/w 2 numeric variabale

visualsation of correlation coffiecient or relationship of 2 numeric data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
temp= [14.2,16.4,11.9,15.2,18.5,22.1,19.4]
icecream_sales=[215,325,185,332,406,522,412]
tea_sales=[80,65,89,70,56,45,43]
```

# 1. Draw A Scatter Plot With Title, Xlabel, And YLabel


```python
plt.scatter(temp,icecream_sales)
plt.title("temp vs icecream_sales")
plt.xlabel("Temp")
plt.ylabel("Icecream sales")
plt.show()
```

# 2. Change Size of The-Dots


```python
plt.scatter(temp,icecream_sales,s=100)
plt.title("temp vs icecream_sales")
plt.xlabel("Temp")
plt.ylabel("Icecream sales")
plt.show()
```

# 3. Change Color of The-Dots


```python
plt.scatter(temp,icecream_sales,
            s=100,
            c="red")
plt.title("temp vs icecream_sales")
plt.xlabel("Temp")
plt.ylabel("Icecream sales")
plt.show()
```

# 4. Change The Marker


```python
plt.scatter(temp,icecream_sales,
            s=100,
            c="red",
            marker=7,
            
           )
plt.title("temp vs icecream_sales")
plt.xlabel("Temp")
plt.ylabel("Icecream sales")
plt.show()
```

# 5. Change The Edge Color of The Marker Also change Line Width


```python
plt.scatter(temp,icecream_sales,
            s=100,
            c="green",
            marker=7,
            facecolor="k",
            linewidths=4,
            
           )
plt.title("temp vs icecream_sales")
plt.xlabel("Temp")
plt.ylabel("Icecream sales")
plt.show()
```

# 6. Change Alpha Parameter


```python
plt.scatter(temp,icecream_sales,
            s=100,
            c="green", # change marker color
            marker="o", # change marker style/design
            facecolor="y", # change marker border color
            linewidths=10, # add border in marker
            alpha=0.75,    # 0 - 100 transparency scale for marker   
            
           )
plt.title("temp vs icecream_sales")
plt.xlabel("Temp")
plt.ylabel("Icecream sales")
plt.show()
```

# 7. Set Different Colors For Different Dots


```python
color=[x for x in range(11,18)]
color
```

# 9. Add Colorbar 


```python
plt.scatter(temp,icecream_sales,
            s=100,
            c=color, # change marker color
            marker="o", # change marker style/design
            facecolor="b", # change marker border color
            linewidths=2, # add border in marker
            alpha=1,    # 0 - 100 transparency scale for marker  
            cmap="hot" # matplotlib buit in color using
            
           )
plt.title("temp vs icecream_sales")
plt.xlabel("Temp")
plt.ylabel("Icecream sales")
plt.colorbar() # add color bar
plt.show()
```

# 10. Customize The Limit of Colorbar 


```python
plt.scatter(temp,icecream_sales,
            s=100,
            c=color, # change marker color
            marker="o", # change marker style/design
            facecolor="b", # change marker border color
            linewidths=2, # add border in marker
            alpha=1,    # 0 - 100 transparency scale for marker  
            cmap="hot", # matplotlib buit in color using
            vmin=10, # min value of color bar
            vmax=20, # max value of color bar
            
           )
plt.title("temp vs icecream_sales")
plt.xlabel("Temp")
plt.ylabel("Icecream sales")
plt.colorbar().set_label("Temp vs Icecream_Sales") # add color bar & give name/title
plt.show()
```

# 12. Draw Two Scatter Plots In One Figure


```python
plt.scatter(temp,icecream_sales,
            s=100,
            c="lightpink", # change marker color
            marker="o", # change marker style/design
            edgecolor="red", # change marker border color
            linewidths=2, # add border in marker
            alpha=1,    # 0 - 100 transparency scale for marker  
           )

plt.scatter(temp,tea_sales,
            s=100,
            c="lightgreen", # change marker color
            marker="o", # change marker style/design
            edgecolor="g", # change marker border color
            linewidths=2, # add border in marker
            alpha=1,    # 0 - 100 transparency scale for marker  
           )


plt.title("temp vs icecream_sales")
plt.xlabel("Temp")
plt.ylabel("Icecream sales")
plt.show()
```

# 13. Add Legend To The Plot


```python
plt.scatter(temp,icecream_sales,
            s=100,
            c="lightpink", # change marker color
            marker="o", # change marker style/design
            edgecolor="red", # change marker border color
            linewidths=2, # add border in marker
            alpha=1,    # 0 - 100 transparency scale for marker 
            label="icecream_sales"
           )

plt.scatter(temp,tea_sales,
            s=100,
            c="lightgreen", # change marker color
            marker="o", # change marker style/design
            edgecolor="g", # change marker border color
            linewidths=2, # add border in marker
            alpha=1,    # 0 - 100 transparency scale for marker 
            label="tea_sales"
           )


plt.title("temp vs icecream_sales")
plt.xlabel("Temp")
plt.ylabel("Icecream sales")
plt.legend()
plt.show()
```
