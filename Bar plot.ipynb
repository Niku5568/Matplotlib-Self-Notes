```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# Bar Plot

```python
grades = ['A','B','C','D']
test1=[3,13,10,4]
test2=[5,14,6,5]
test3=[7,15,5,3]
```

# draw bar plot

```python
plt.bar(grades,test1)
plt.title("Bar Plot",fontsize=15)
plt.xlabel("Grade")
plt.ylabel("No of Student")
plt.show()
```

# allign bar into center and set width

```python
plt.bar(grades,test1,align="center",width=0.5)
plt.title("Bar Plot",fontsize=15)
plt.xlabel("Grade")
plt.ylabel("No of Student")
plt.show()
```

# change color of the bar and edge color of the bars

```python
plt.bar(grades,test1,align="center",width=0.5,color=["r","k","g","b"],edgecolor=["b","g","k","y"])
plt.title("Bar Plot",fontsize=15)
plt.xlabel("Grade")
plt.ylabel("No of Student")
plt.show()
```

```python
plt.bar(grades,test1,align="center",width=0.5,color="grey",edgecolor="k")
plt.title("Bar Plot",fontsize=15)
plt.xlabel("Grade")
plt.ylabel("No of Student")
plt.show()
```

# change line styles & line width of bars

```python
plt.bar(grades,test1,align="center",width=0.5,color="brown",edgecolor="k",ls="--",lw=2)
plt.title("Bar Plot",fontsize=15)
plt.xlabel("Grade")
plt.ylabel("No of Student")
plt.show()
```

# draw bar plot for test 1 and test 2

```python
plt.bar(grades,test1,align="center",width=0.5,color="brown",edgecolor="k",ls="--",lw=2)
plt.bar(grades,test2,align="center",width=0.5,color="blue",edgecolor="k",ls="--",lw=2)

plt.title("Bar Plot",fontsize=15)
plt.xlabel("Grade")
plt.ylabel("No of Student")
plt.show()
```

```python
index=np.arange(len(grades))
index
```

```python
plt.bar(index,test1,align="center",width=0.3,color="brown",edgecolor="k",ls="--",lw=1,label="Test1")
plt.bar(index+0.3,test2,align="center",width=0.3,color="blue",edgecolor="k",ls="--",lw=1,label="Test2")
plt.title("Bar Plot",fontsize=15)
plt.xlabel("Grade")
plt.ylabel("No of Student")
plt.legend()
plt.show()
```

# change x axis float value into grades and align it into center

```python
plt.bar(index,test1,align="center",width=0.3,color="brown",edgecolor="k",ls="--",lw=1,label="Test1")
plt.bar(index+0.3,test2,align="center",width=0.3,color="blue",edgecolor="k",ls="--",lw=1,label="Test2")
plt.title("Bar Plot",fontsize=15)
plt.xlabel("Grade")
plt.ylabel("No of Student")
plt.xticks(index+0.2,grades)
plt.legend()
plt.show()
```

# draw bar plot for test 1 and test 2 and test 3

```python
plt.bar(index,test1,align="center",width=0.3,color="r",edgecolor="k",ls="--",lw=1,label="Test1")
plt.bar(index+0.3,test2,align="center",width=0.3,color="blue",edgecolor="k",ls="--",lw=1,label="Test2")
plt.bar(index+0.6,test3,align="center",width=0.3,color="g",edgecolor="k",ls="--",lw=1,label="Test3")
plt.title("Bar Plot",fontsize=15)
plt.xlabel("Grade")
plt.ylabel("No of Student")
plt.xticks(index+0.3,grades,rotation=60)
plt.legend()
plt.show()
```

# change fig color & size

```python
plt.figure(figsize=(10,3),facecolor="lightgreen",edgecolor="green",linewidth=5)
plt.bar(index,test1,align="center",width=0.3,color="r",edgecolor="k",ls="--",lw=1,label="Test1")
plt.bar(index+0.3,test2,align="center",width=0.3,color="blue",edgecolor="k",ls="--",lw=1,label="Test2")
plt.bar(index+0.6,test3,align="center",width=0.3,color="g",edgecolor="k",ls="--",lw=1,label="Test3")
plt.title("Bar Plot",fontsize=15)
plt.xlabel("Grade")
plt.ylabel("No of Student")
plt.xticks(index+0.3,grades,rotation=60)
plt.legend()
plt.show()
```

# save graph

```python
plt.figure(figsize=(10,3),facecolor="lightgreen",edgecolor="green",linewidth=5)
plt.bar(index,test1,align="center",width=0.3,color="r",edgecolor="k",ls="--",lw=1,label="Test1")
plt.bar(index+0.3,test2,align="center",width=0.3,color="blue",edgecolor="k",ls="--",lw=1,label="Test2")
plt.bar(index+0.6,test3,align="center",width=0.3,color="g",edgecolor="k",ls="--",lw=1,label="Test3")
plt.title("Bar Plot",fontsize=15)
plt.xlabel("Grade")
plt.ylabel("No of Student")
plt.xticks(index+0.3,grades,rotation=90)
plt.legend()
plt.savefig("barplot.png")
plt.show()
```
