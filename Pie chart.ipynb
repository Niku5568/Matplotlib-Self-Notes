```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# Pie chart

```python
movie = ['Comedy','Action','Romance','Drama']
votes=[5,6,3,1]
```

# create Pie chart

```python
plt.pie(votes)
plt.show()
```

# Give Title, Label (Also change label distance) and legend

```python
plt.pie(votes,labels=movie,autopct="%.f%%",shadow=True,labeldistance=1.1)
plt.title("Type of Movies")
plt.legend(loc=4)
plt.show()
```

# pull out action and drama wedges

```python
plt.pie(votes,labels=movie,autopct="%.f%%",labeldistance=1.1,explode=[0,0.05,0,0.1])
plt.title("Type of Movies")
plt.legend(loc=4)
plt.show()
```

# change the start angle to 90 degree

```python
plt.pie(votes,labels=movie,autopct="%.f%%",labeldistance=1.1,explode=[0,0.05,0,0.1],startangle=90)
plt.title("Type of Movies")
plt.legend(loc=4)
plt.show()
```

# set the color of each wedges

```python
plt.pie(votes,labels=movie,autopct="%.f%%",labeldistance=1.1,explode=[0,0.05,0,0.1],startangle=90,colors=["c","grey","m","orange"])
plt.title("Type of Movies")
plt.legend(loc=4)
plt.show()
```

# display percentage and change distance inside each wedges

```python
plt.pie(votes,labels=movie,autopct="%.f%%",labeldistance=1.1,explode=[0,0.05,0,0.1],startangle=90,colors=["c","grey","m","orange"],pctdistance=0.6)
plt.title("Type of Movies")
plt.legend(loc=4)
plt.show()
```

# change radius of the pie chart

```python
plt.pie(votes,labels=movie,autopct="%.f%%",labeldistance=1.1,explode=[0,0.05,0,0.1],startangle=90,colors=["c","grey","m","orange"],
        pctdistance=0.6,radius=1.2)
plt.title("Type of Movies",y=1.1)
plt.legend(loc=3)
plt.show()
```

# set label size of the pie chart

```python
plt.pie(votes,
        labels=movie, # add name to each slice
        autopct="%.f%%", # add percentage 
        labeldistance=1.1, # label distance
        explode=[0,0.05,0,0.1], # pullout particular slice
        startangle=90, # change slicing angle
        colors=["c","grey","m","orange"], # change color of each slice 
        pctdistance=0.6, # percentage distance
        radius=1.2,  # enlarge pie chart 
        textprops={"fontsize":12} # increase label text size
        
       
       )
plt.title("Type of Movies",y=1.1)
plt.legend(loc=3)
plt.show()
```

# set border of the wedges/slice

```python
plt.pie(votes,
        labels=movie, # add name to each slice
        autopct="%.f%%", # add percentage 
        labeldistance=1.1, # label distance
        explode=[0,0.05,0,0.1], # pullout particular slice
        startangle=90, # change slicing angle
        colors=["c","grey","m","orange"], # change color of each slice 
        pctdistance=0.6, # percentage distance
        radius=1.2,  # enlarge pie chart 
        textprops={"fontsize":12}, # increase label text size 
        wedgeprops={"edgecolor":"b",
                   "linewidth":1,
                   "linestyle":":"}, # add border color and widthsize
        
       
       )
plt.title("Type of Movies",y=1.1)
plt.legend(loc=3)
plt.show()
```

# add shadow

```python
plt.pie(votes,
        labels=movie, # add name to each slice
        autopct="%.f%%", # add percentage 
        labeldistance=1.1, # label distance
        explode=[0,0.05,0,0.1], # pullout particular slice
        startangle=90, # change slicing angle
        colors=["c","grey","m","orange"], # change color of each slice 
        pctdistance=0.6, # percentage distance
        radius=1.2,  # enlarge pie chart 
        textprops={"fontsize":12}, # increase label text size 
        wedgeprops={"edgecolor":"b",
                   "linewidth":1,
                   "linestyle":":"}, # add border color and widthsize
        shadow=True
       
       )
plt.title("Type of Movies",y=1.1)
plt.legend(loc=3)
plt.show()
```

# change figure color and size

```python
plt.figure(figsize=(10,5),edgecolor="green",linewidth=3,facecolor="pink")
plt.pie(votes,
        labels=movie, # add name to each slice
        autopct="%.f%%", # add percentage 
        labeldistance=1.1, # label distance
        explode=[0,0.05,0,0.1], # pullout particular slice
        startangle=90, # change slicing angle
        colors=["c","grey","m","orange"], # change color of each slice 
        pctdistance=0.6, # percentage distance
        radius=1.2,  # enlarge pie chart 
        textprops={"fontsize":12}, # increase label text size 
        wedgeprops={"edgecolor":"lightblue",
                   "linewidth":1,
                   "linestyle":"-"}, # add border color and widthsize
)
plt.title("Type of Movies",y=1.1,color="k")
plt.legend(loc=3)
plt.show()
```
