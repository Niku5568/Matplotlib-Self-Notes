# 3D surface

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y) # used to create grids in numpy to plot surface chart
```

```python
# calculate z by using mesgrid
z= xx**2 + yy**2
z.shape

fig=plt.figure(figsize=(10,5))
ax=fig.add_subplot(projection='3d')
sp=ax.plot_surface(xx,yy,z,cmap="viridis")

fig.colorbar(sp)
plt.show()
```

```python
z= np.sin(xx)+np.cos(yy)
z.shape

fig=plt.figure(figsize=(10,5))
ax=fig.add_subplot(projection='3d')
sp=ax.plot_surface(xx,yy,z,cmap="cool")

fig.colorbar(sp)
plt.show()
```

# contour plot

```python
z= xx**2 + yy**2
z.shape

fig=plt.figure(figsize=(10,5))
ax=fig.add_subplot()
sp=ax.contour(xx,yy,z,cmap="viridis")

fig.colorbar(sp)
plt.show()
```
